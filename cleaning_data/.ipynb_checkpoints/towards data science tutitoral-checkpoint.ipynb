```python
# excerise from https://towardsdatascience.com/data-cleaning-with-python-and-pandas-detecting-missing-values-3e9c6ebcf78b

import pandas as pd
import numpy as np
```

```python
# in read_csv, na_values= sets a particular value to a NaN
missing_value = ['n/a', "na", "--"]
df = pd.read_csv('https://raw.githubusercontent.com/dataoptimal/posts/master/data%20cleaning%20with%20python%20and%20pandas/property%20data.csv', na_values=missing_value)
```

```python
df.head(20)
```

```python
df.rename(columns={'PID': 'POID'}, inplace=True)
```

```python
df
```

```python
df.isnull().mean()
```

```python
cnt = 0
for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        df.loc[cnt,'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
```

```python
data = {'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'year': [2012, 2012, 2013, 2014, 2014], 
        'reports': [4, 24, 31, 2, 3]}
df = pd.DataFrame(data, index = ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'])
df
```

```python
# create a variable
next_year = []

for row in df['year']:
    next_year.append(row+1)
    
df['next_year'] = next_year
```

```python
df['last_year'] = [row-1 for row in df['year']]
```
